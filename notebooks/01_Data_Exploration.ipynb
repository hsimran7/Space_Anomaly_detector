```python
df = df.reset_index()
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df['normalized_flux'] = scaler.fit_transform(df[['pdcsap_flux']])

print(df[['pdcsap_flux', 'normalized_flux']].head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.plot(df.index, df['normalized_flux'])

plt.xlabel("Time")
plt.ylabel("Normalized Flux")
plt.title("Normalized Kepler Light Curve")

plt.show()
```

```python
import numpy as np

sequence_length = 100

X = []

data = df['normalized_flux'].values

for i in range(len(data) - sequence_length):
    X.append(data[i:i + sequence_length])

X = np.array(X)

print(X.shape)
```

```python
X = X.reshape((X.shape[0], X.shape[1], 1))

print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(
    X,
    test_size=0.2,
    random_state=42,
    shuffle=False
)

print(X_train.shape)
print(X_test.shape)
```
